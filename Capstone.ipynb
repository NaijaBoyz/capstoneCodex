```python
import pandas as pd

# Load datasets
path_prefix = "C:\\Users\\gbemi\\Desktop\\Capstone\\"
weather_events = pd.read_csv(path_prefix + "WeatherEvents_Jan2016-Dec2022.csv")
data_2016 = pd.read_csv(path_prefix + "2016.csv")
data_2017 = pd.read_csv(path_prefix + "2017.csv")
data_2018 = pd.read_csv(path_prefix + "2018.csv")
```

```python
delay_df = pd.concat([data_2016, data_2017, data_2018], ignore_index=True)
```

```python
delay_df.tail()
```

```python
weather_delay_counts = delay_df['WEATHER_DELAY'].value_counts()
print(weather_delay_counts)
```

```python
# Remove rows where WEATHER_DELAY is NaN or 0
delay_df = delay_df[delay_df['WEATHER_DELAY'].notna() & (delay_df['WEATHER_DELAY'] != 0)]
```
